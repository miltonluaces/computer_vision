### Toy

```python
snapshot_pp_toy = cv2.imread(img_path + 'ocr/snapshot_pp_toy.png')
plt.figure(figsize=(15,10))
plt.imshow(snapshot_pp_toy);                        
```

```python
txt = pytesseract.image_to_string(snapshot_pp_toy)
lines = txt.split('\n')
lines
```

```python
filt_lines = []
for line in lines:
    if(line != '\x0c'): filt_lines.append(line)
        
filt_lines
```

```python
lines_txt = '\n'.join(filt_lines) + '\n'
```

```python
ocr_txt = open("ocr2.txt","w") 
ocr_txt.write(lines_txt) 
ocr_txt.close()
```
